12. Отсортируйте матрицу $A$ по 3-ому столбцу, т.е. поменяйте местами строки матрицы так, чтобы 3-й столбец оказался отсортированным. Используйте для этого slicing + argsort + indexing.

```python
col = A[:, 3]
sorted = np.argsort(A[:, 3])
temp = np.zeros((3,))
for i in range(3):
    temp[i] = col[sorted[i]]
A[:, 3] = temp
A
```
